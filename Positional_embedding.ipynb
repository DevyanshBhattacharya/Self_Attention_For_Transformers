```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import os
```

```python
def positional_encoding(length:int ,depth: int):
  depth=depth/2

  positions= np.arange(length)[:,np.newaxis] #Input (matrix) sequence of 1
  depths= np.arange(depth)[np.newaxis,:]/depth #Input (matrix) sequence of 2

  angle_rads = 1/(10000**depths)
  depths = positions * angle_rads

  pos_encoding = np.concatenate([np.sin(angle_rads),np.cos(angle_rads)], axis=-1)

  return tf.cast(pos_encoding, dtype=tf.float32)
```

```python
pos_encoding = positional_encoding(2048,512)
print(pos_encoding.shape)

plt.pcolormesh(pos_encoding.numpy().T, cmap='RdBu')
plt.ylabel('Depth')
plt.xlabel('Position')
plt.colorbar()
plt.show()
```
